# TCGA-OV CAR Panel Scaffold Notebook

This notebook demonstrates the **metadata-only CAR benchmarking scaffold** included with the public release.

It is intentionally limited to:
- manifest inspection
- architecture metadata review
- tool readiness auditing
- command-template inspection

It does **not** include engineered CAR construct FASTA content.


```python
import os
from pathlib import Path
import pandas as pd

DATA_ROOT = Path("/kaggle/input/tcga-ov-multiomics-network-derived-results")
assert DATA_ROOT.exists(), f"Dataset path not found: {DATA_ROOT}"
DATA_ROOT
```
