```python
file_inventory = pd.read_csv(tables / "external_validation_file_inventory.csv")
cart_qc = pd.read_csv(tables / "cart_direct_benchmark_qc.csv")
motif = pd.read_csv(tables / "cart_motif_benchmark.csv")
ready = pd.read_csv(tables / "cart_reference_alignment_readiness.csv")
display(file_inventory)
display(cart_qc)
```

## Host-Alignment Readiness


```python
display(ready)
tools = ready[ready["component"].str.startswith("tool:")]
display(tools)
print("Host-alignment-ready tools available:")
print(tools[["component", "status", "detail"]].to_string(index=False))
```
